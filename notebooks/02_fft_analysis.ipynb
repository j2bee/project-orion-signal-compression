# 02 — FFT Analysis

Detailed FFT compression parameter analysis.

```python
import sys
sys.path.insert(0, '..')
import numpy as np
import matplotlib.pyplot as plt
from src.data_loader.loader import load_signal
from src.generate_synthetic import create_synthetic_dataset
from src.compression.fft_compression import compress_fft, decompress_fft
from src.metrics.mse import mse
from src.metrics.snr import snr_db
from src.metrics.compression_ratio import compression_ratio
%matplotlib inline
```

```python
path = create_synthetic_dataset()
time, signal, meta = load_signal(path)
rates = np.linspace(0.01, 1.0, 20)
mse_vals, snr_vals, ratios = [], [], []
for r in rates:
    c = compress_fft(signal, r)
    recon = decompress_fft(c)
    mse_vals.append(mse(signal, recon))
    snr_vals.append(snr_db(signal, recon))
    ratios.append(compression_ratio(signal, c))
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].plot(rates*100, mse_vals); axes[0].set_xlabel('Keep %'); axes[0].set_ylabel('MSE'); axes[0].set_title('MSE vs Keep %')
axes[1].plot(rates*100, snr_vals); axes[1].set_xlabel('Keep %'); axes[1].set_ylabel('SNR (dB)'); axes[1].set_title('SNR vs Keep %')
axes[2].plot(rates*100, ratios); axes[2].set_xlabel('Keep %'); axes[2].set_ylabel('Ratio'); axes[2].set_title('Compression Ratio vs Keep %')
plt.tight_layout();
```
