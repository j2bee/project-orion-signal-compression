# 04 — Noise Analysis

Study how different noise models affect compression quality.

```python
import sys
sys.path.insert(0, '..')
import numpy as np
import matplotlib.pyplot as plt
from src.data_loader.loader import load_signal
from src.generate_synthetic import create_synthetic_dataset
from src.preprocessing.noise import add_gaussian_noise, add_impulse_noise, add_sensor_drift
from src.compression.fft_compression import compress_fft, decompress_fft
from src.metrics.snr import snr_db
from src.metrics.mse import mse
%matplotlib inline
```

```python
path = create_synthetic_dataset()
time, signal, meta = load_signal(path)
noise_levels = [0.01, 0.05, 0.10, 0.20, 0.50]
print(f"{'Noise Level':>12} {'Pre-comp SNR':>14} {'Post-comp SNR':>14} {'MSE':>10}")
for level in noise_levels:
    noisy = add_gaussian_noise(signal, level, seed=42)
    comp = compress_fft(noisy, 0.1)
    recon = decompress_fft(comp)
    print(f"{level:>11.0%} {snr_db(signal, noisy):>14.1f} {snr_db(signal, recon):>14.1f} {mse(signal, recon):>10.6f}")
```
