Compare FFT, wavelet, and quantization methods.

```python
import sys
sys.path.insert(0, '..')
from src.data_loader.loader import load_signal
from src.generate_synthetic import create_synthetic_dataset
from src.compression.fft_compression import compress_fft, decompress_fft
from src.compression.wavelet_compression import compress_wavelet, decompress_wavelet
from src.compression.quantization import compress_quantization, decompress_quantization
from src.metrics.mse import mse
from src.metrics.snr import snr_db
from src.metrics.compression_ratio import compression_ratio
from src.visualization.plots import plot_comparison
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = create_synthetic_dataset()
time, signal, meta = load_signal(path)
methods = {
    'FFT 10%': (compress_fft(signal, 0.1), decompress_fft),
    'Wavelet db4 10%': (compress_wavelet(signal, 'db4', 0.1), decompress_wavelet),
    'Quantization 8-bit': (compress_quantization(signal, 8), decompress_quantization),
}
for name, (comp, decomp) in methods.items():
    recon = decomp(comp)
    print(f"{name:25s} MSE={mse(signal,recon):.6f} SNR={snr_db(signal,recon):.1f}dB Ratio={compression_ratio(signal,comp):.1f}x")
```
